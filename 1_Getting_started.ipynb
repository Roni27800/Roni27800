```python
def two_numbers(num1, num2):
    # Check if both inputs are digits
    if num1.isdigit() and num2.isdigit():
        # Convert to integers after validation
        num1 = int(num1)
        num2 = int(num2)

        sum_result = num1 + num2
        diff = num1 - num2

        print(f"Sum: {sum_result}")
        print(f"Difference: {diff}")
    else:
        print("Both inputs must be numbers.")

# Get input from the user as strings
two_numbers(input('Enter the first number: '), input('Enter the second number: '))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

order_details = pd.read_csv('/content/sample_data/order_details.csv')
orders = pd.read_csv('/content/sample_data/orders.csv')
pizza_types = pd.read_json('/content/sample_data/pizza_types_json.txt')
pizzas = pd.read_csv('/content/sample_data/pizzas.csv')
```

```python
full_table = pizza_types.merge(pizzas, how = 'outer').merge(order_details, how = 'outer').merge(orders, how = 'outer')
full_table
```

```python
df = pd.DataFrame(full_table)
```

```python
print(len(df['name'].drop_duplicates()))
```

```python
df['name'].drop_duplicates().reset_index()
```

```python
df['total'] = df.price * df.quantity
df.total
```

```python
new_df = df[['pizza_type_id', 'name', 'total']]
```

```python
new_df
```

```python
grouped_df = df.groupby(['pizza_type_id', 'name'])['total'].sum()
new_pizza = grouped_df.reset_index().sort_values(by = 'total', ascending = False)
```

```python
# Rank the 'total' column in descending order and add it as a new 'rank' column
new_pizza['rank'] = new_pizza['total'].rank(ascending=False, method='dense').astype(int)
```

```python
top_3_pizzas = new_pizza.head(3)
```

```python
pizzas = top_3_pizzas
```

```python
pizzas
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_xlabel('Pizza Type')
ax.set_ylabel('Sales')
ax.set_title('Sales by Pizza Type')


ax.bar(pizzas.name, pizzas.total)

max_total = pizzas.total.max()
min_total = pizzas.total.min()
ax.set_ylim(min_total - 1500, max_total + 1000)

plt.show()
```
